```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# read in the data
df = pd.read_csv(r'C:\Users\xxxx\Documents\Portfolio\Project_x\movies.csv')
```

```python
# Now let's take a look at the data

df.head()
```

```python
# is there anything missing?

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#change data type

df['budget'] = df['budget'].astype("Int64")
df['gross'] = df['gross'].astype("Int64")

df.head()
```

```python
#sort the data
df= df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
# visualize the correlation
correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#convert the strings like company name to a number to be able to analyze correlation

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes  #it gives the random numerization
        
df_numerized.head()
```

```python
# visualize the correlation
correlation_matrix = df_numerized.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr()
```

```python
# to view the correlation easier
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
hihg_corr = sorted_pairs[(sorted_pairs) > 0.5]
hihg_corr 
```
